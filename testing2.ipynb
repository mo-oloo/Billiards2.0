```python
part = pt(-1, -1, 1, 3)

tot_height = 6
tot_width = 6
bound = bd(tot_width, tot_height)

scatterer0 = sc(0, 0, 1)
scatterer1 = sc(3, 3, 1)
scatterer2 = sc(3, -3, 1)
scatterer3 = sc(-3, 3, 1)
scatterer4 = sc(-3, -3, 1)

scatterers = [scatterer0, scatterer1, scatterer2, scatterer3, scatterer4]

billiards = BS(particle = part, 
                boundary = bound, 
                scatterers = scatterers)
```

```python
n = 10000
collisions = billiards.run_simulation(n)
print('length of dataset: ', len(collisions))
```
